```python
import requests
```

```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
# Get all network information
x = requests.get('http://api.citybik.es/v2/networks')
x.status_code
```

```python
networks = json.loads(x.content)
networks
```

```python
# Only need some of the fields 
n_limited = requests.get('http://api.citybik.es/v2/networks', params={'fields':'id,name'})
networks_limited_json = json.loads(n_limited.content)

networks_df = pd.DataFrame.from_dict(networks_limited_json.pop('networks'))
networks_df
```

```python
# Pull Station data for DC's Bike Share, Capital Bikeshare
c = requests.get('http://api.citybik.es/v2/networks/capital-bikeshare', params={'fields':'stations'})
cabi = json.loads(c.content)
# cabi
cabi_stations = pd.DataFrame.from_dict(cabi.pop('network').pop('stations'))
cabi_stations
```

```python
# stations with the most bikes
cabi_stations.sort_values('free_bikes', ascending=False).head(10)
```

```python
# stations with the most parking spots
cabi_stations.sort_values('empty_slots', ascending=False).head(10)
```
